#Country Rankings

##Imports and reading CSV's

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

```python
gender_development_df = pd.read_csv("/content/gender_development.csv")
gender_inequality_df = pd.read_csv("/content/gender_inequality.csv")
human_development_df = pd.read_csv("/content/human_development.csv")
multi_poverty_df = pd.read_csv("/content/multidimensional_poverty.csv")

poverty_df = pd.read_csv("/content/poverty-share-on-less-than-30-per-day.csv")

internet_df = pd.read_csv("/content/Internet Speed 2022.csv")
innovation_df = pd.read_csv("/content/Analysis_2023abr.29184621.csv")

happiness_df = pd.read_csv("/content/world-happiness-report-2021.csv")

pollution_df = pd.read_csv("/content/global air pollution dataset.csv")
death_pollution_df = pd.read_csv("/content/death-rates-from-air-pollution.csv")

emissions_df = pd.read_csv("/content/GCB2022v27_MtCO2_flat.csv")

health_df = pd.read_csv("/content/country_health_indicators_v3.csv")

gpd_capita_df = pd.read_csv("/content/gdp_per_capita.csv")
gdp_capita_growth_df = pd.read_csv("/content/gdp_per_capita_growth.csv")
```

##Sorting and Wrangling

###Social Indicators

####Happiness

```python
happiness_final = happiness_df[['Country name', 'Ladder score']]
happiness_final = happiness_final.rename(columns = {'Country name': 'Country', 'Ladder score': 'Happiness Ladder Score'})
happiness_final
```

####Gender Development

```python
gender_development_final = gender_development_df[['Country', 'Gender Development Index (GDI)']]
gender_development_final = gender_development_final.rename(columns = {'Country name': 'Country', 'Gender Development Index (GDI)': 'Gender Development Index'})
gender_development_final
```

#### Expected Years of Education

```python
expected_education_final = human_development_df[['Country', 'Expected Years of Education']]
expected_education_final
```

#### Freedom to make choices

```python
freedom_final = happiness_df[['Country name','Freedom to make life choices']]
freedom_final = freedom_final.rename(columns = {'Country name': 'Country', 'Freedom to make life choices':'Freedom to Make Life Choices'})
freedom_final
```

#### Social Support

```python
social_support_final = happiness_df[['Country name','Social support']]
social_support_final = social_support_final.rename(columns = {'Country name': 'Country', 'Social support':'Social Support'})
social_support_final
```

#### Generosity

```python
generosity_final = happiness_df[['Country name','Generosity']]
generosity_final = generosity_final.rename(columns = {'Country name': 'Country'})
generosity_final
```

####FINAL Social Development DF

```python
temp_df = pd.merge(happiness_final, gender_development_final, how = 'right', on = 'Country')
temp_df = pd.merge(temp_df, expected_education_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, freedom_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, social_support_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, generosity_final, how = 'left', on = 'Country')
social_development_df = temp_df
imputer = KNNImputer(n_neighbors=5)
ml_df = social_development_df.drop(['Country'], axis = 1)
columns = ml_df.columns
ml_df = ml_df.replace("..", np.nan)
ml_df = imputer.fit_transform(ml_df)
social_indicators = pd.DataFrame(ml_df, columns = columns)
social_indicators['Country'] = social_development_df['Country']
social_indicators
```

###Health Indicators

```python
health_df.columns
```

```python
health_final = health_df[['Country_Region', 'life expectancy at birth', 'medical_doctors_per10k', 'Nutritional deficiencies (%)','Cancers (%)',
                          'Diabetes, blood, & endocrine diseases (%)', 'HIV/AIDS and tuberculosis (%)']]
health_final = health_final.rename(columns = {"Country_Region": "Country", "life expectancy at birth": "Life Expectancy",
                                              'medical_doctors_per10k':'Doctors per 10k people', 'Nutritional deficiencies (%)': 'Nutritional Deficiencies',
                                              'Cancers (%)':'Cancer %', 'Diabetes, blood, & endocrine diseases (%)': 'Diabetes %', 'HIV/AIDS and tuberculosis (%)': 'HIV/AIDS and Tuberculosis %'})
#health_final.sort_values(by= "Life Expectancy", ascending= False)
ml_df = health_final.drop(['Country'], axis = 1)
columns = ml_df.columns
#ml_df = ml_df.replace("..", np.nan)
ml_df = imputer.fit_transform(ml_df)
health_indicators = pd.DataFrame(ml_df, columns = columns)
health_indicators['Country'] =health_final['Country']
health_indicators
```

###Economic Indicators

####GDP per Capita

```python
gpd_capita_df
```

```python
gdp_capita_final = gpd_capita_df[['Country Name', '2020']]
gdp_capita_final = gdp_capita_final.rename(columns = {'Country Name': "Country", '2020':'GDP Per Capita'})
#gpd_capita_final.sort_values(by = "GDP Per Capita", ascending = False)
gdp_capita_final
```

#### GDP per Capita growth Rate

```python
gdp_capita_growth_df
```

```python
gdp_capita_growth_final = gdp_capita_growth_df[['Country Name', '2020']]
gdp_capita_growth_final = gdp_capita_growth_final.rename(columns = {'Country Name': "Country", '2020':'GDP Per Capita Growth Rate'})
gdp_capita_growth_final
```

####% Living on Less than $30 Per Day

```python
poverty_final = poverty_df[poverty_df['Year'] == 2018]
poverty_final = poverty_final[['Entity', '$30.00 per day - share of population below poverty line']]
poverty_final = poverty_final.rename(columns = {'$30.00 per day - share of population below poverty line' : '% Living on Less than $30 Per Day', 'Entity': 'Country'})
poverty_final
```

####Internet

```python
internet_final = internet_df[['country', 'broadband']]
internet_final = internet_final.rename(columns = {'country': 'Country', 'broadband': 'Internet Speed'})
internet_final
```

####Multi-Dimensional Poverty Index

```python
mdpi_final = multi_poverty_df[["Country","Multidimensional Poverty Index (MPI, 2010)"]]
mdpi_final = mdpi_final.rename(columns = {'Multidimensional Poverty Index (MPI, 2010)': 'Multi-Dimensional Poverty Index'})
mdpi_final
```

####Income Inequality

```python
income_inequality_final = gender_inequality_df[['Country', 'Gender Inequality Index (GII)']]
income_inequality_final = income_inequality_final.rename(columns = {'Country name': 'Country', 'Gender Inequality Index (GII)': 'Income Inequality'})
income_inequality_final
```

####FINAL Economic Indicators

```python
temp_df = pd.merge(gdp_capita_final ,gdp_capita_growth_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, poverty_final, how = 'right', on = 'Country')
temp_df = pd.merge(temp_df, internet_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, mdpi_final, how = 'left', on = 'Country')
temp_df = pd.merge(temp_df, income_inequality_final, how = 'left', on = 'Country')
final_wealth_df = temp_df
imputer = KNNImputer(n_neighbors=5)
ml_df = final_wealth_df.drop(['Country'], axis = 1)
columns = ml_df.columns
ml_df = ml_df.replace("..", np.nan)
ml_df = imputer.fit_transform(ml_df)
economic_indicators = pd.DataFrame(ml_df, columns = columns)
economic_indicators['Country'] = final_wealth_df['Country']
economic_indicators
```

###Pollution Indicators

####AQI Value

```python
pollution_df
```

```python
aqi_final = pollution_df.groupby("Country").mean().reset_index()
aqi_final = aqi_final[['Country', 'AQI Value']]
aqi_final = aqi_final.rename(columns = {'AQI Value': 'Air Quality Index'})
aqi_final
```

####Pollution Deaths per Million

```python
death_pollution_df
```

```python
death_pollution_final = death_pollution_df.groupby("Entity").mean().reset_index()
death_pollution_final = death_pollution_final[['Entity', 'Air pollution (total) (deaths per 100,000)']]
death_pollution_final = death_pollution_final.rename(columns = {'Entity': 'Country','Air pollution (total) (deaths per 100,000)': 'Pollution Deaths per 100,000'})
death_pollution_final
```

####CO2 Emissions 
(Should we use per capita here?)

```python
emissions_df
```

```python
emissions_final = emissions_df[emissions_df['Year'] == 2020]
emissions_final = emissions_final[['Country', 'Coal','Oil','Gas','Total']]
emissions_final = emissions_final.rename(columns = {'Coal':'CO2 from Coal', 'Oil':'CO2 from Oil','Gas':'CO2 from Gas','Total': 'CO2 Emissions'})
emissions_final
```

####FINAL POLLUTION DF

```python
temp_df = pd.merge(aqi_final, death_pollution_final , how = 'left', on= 'Country')
temp_df = pd.merge(temp_df, emissions_final , how = 'left', on= 'Country')
pollution_emission_df = temp_df
imputer = KNNImputer(n_neighbors=5)
ml_df = pollution_emission_df.drop(['Country'], axis = 1)
columns = ml_df.columns
#ml_df = ml_df.replace("..", np.nan)
ml_df = imputer.fit_transform(ml_df)
pollution_indicators = pd.DataFrame(ml_df, columns = columns)
pollution_indicators['Country'] = pollution_emission_df['Country']
pollution_indicators
```

##Merging!

```python
final = social_indicators.merge(health_indicators, how = 'left',
                                on = 'Country').merge(economic_indicators, how = 'left', 
                                on = 'Country').merge(pollution_indicators, how = 'left',
                                on = 'Country')
imputer = KNNImputer(n_neighbors=5)
ml_df = final.drop(['Country'], axis = 1)
columns = ml_df.columns
#ml_df = ml_df.replace("..", np.nan)
ml_df = imputer.fit_transform(ml_df)
final_df = pd.DataFrame(ml_df, columns = columns)
final_df.insert(0,'Country',final['Country'].tolist())
final_df
```

```python
final_df.to_csv("CountriesRanking.csv")
```
